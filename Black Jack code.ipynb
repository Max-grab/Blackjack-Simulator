```python
def total(turn):
    total=0
    face = ['J', 'K', 'Q']
    for card in turn:
        if card in range(1,11):
            total +=1
        elif card in face:
             total +=1
    else:
        if total > 11: 
            total += 1
        else: total+= 11
    return total
```
